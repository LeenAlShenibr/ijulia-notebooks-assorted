Wigner's original application of random matrix theory was to study the distribution of nuclear energy levels. In this notebook we revisit Wigner's original application using modern data from the [Evaluated Nuclear Structure Data File Database](http://www.nndc.bnl.gov/ensdf/) at the [National Nuclear Data Center](http://www.nndc.bnl.gov).

To analyze a given nuclide, look up the ENSDF data file associated with it in the database (usually you will want the *ADOPTED LEVELS, GAMMAS* entry). Some of these have already been looked up and stored in the `datafiles` dictionary.

```julia
using Compat #For use with both Julia v0.3 and v0.4
using Gadfly #For plotting
using Interact
```

Here we keep track of the URL of the data set for a given nuclide. The `nuclide` variable specifies which nuclide we want to analyze.

```julia
nuclide = "I131"

#Download data
datafiles = @compat Dict(
    #Radiocarbon dating
    "C14"  => "http://www.nndc.bnl.gov/useroutput/AR_0E57A699F47BD7B9AA46AFD8D289942C_13.ens",

    #Medicine
    "Co60" => "http://www.nndc.bnl.gov/useroutput/AR_0E57A699F47BD7B9AA46AFD8D289942C_10.ens",

    "I129" => "http://www.nndc.bnl.gov/useroutput/AR_0E57A699F47BD7B9AA46AFD8D289942C_17.ens",
    "I130" => "http://www.nndc.bnl.gov/useroutput/AR_0E57A699F47BD7B9AA46AFD8D289942C_18.ens",
    "I131" => "http://www.nndc.bnl.gov/useroutput/AR_0E57A699F47BD7B9AA46AFD8D289942C_19.ens",

    #Nuclear waste
    "Cs137" => "http://www.nndc.bnl.gov/useroutput/AR_0E57A699F47BD7B9AA46AFD8D289942C_14.ens",

    #X-ray source for medical imaging
    "Gd151"=> "http://www.nndc.bnl.gov/useroutput/AR_0E57A699F47BD7B9AA46AFD8D289942C_12.ens",

    #Smoke detectors
    "Am241"=> "http://www.nndc.bnl.gov/useroutput/AR_0E57A699F47BD7B9AA46AFD8D289942C_11.ens",

    "Pb206"=> "http://www.nndc.bnl.gov/useroutput/AR_0E57A699F47BD7B9AA46AFD8D289942C_3.ens",
    "Pb207"=> "http://www.nndc.bnl.gov/useroutput/AR_0E57A699F47BD7B9AA46AFD8D289942C_2.ens",
    "Pb208"=> "http://www.nndc.bnl.gov/useroutput/AR_0E57A699F47BD7B9AA46AFD8D289942C_1.ens",
    "Po210"=> "http://www.nndc.bnl.gov/useroutput/AR_0E57A699F47BD7B9AA46AFD8D289942C_16.ens",
    "U235" => "http://www.nndc.bnl.gov/useroutput/AR_09CB581BB90DBC20099CC8348ED33DD1_2.ens",
    "U238" => "http://www.nndc.bnl.gov/useroutput/AR_09CB581BB90DBC20099CC8348ED33DD1_3.ens",
    "Pu240"=> "http://www.nndc.bnl.gov/useroutput/AR_0E57A699F47BD7B9AA46AFD8D289942C_4.ens",
    "Pu241"=> "http://www.nndc.bnl.gov/useroutput/AR_0E57A699F47BD7B9AA46AFD8D289942C_5.ens",
)

fn = nuclide * ".ens"
isfile(fn) || download(datafiles[nuclide], fn)

#Parse ENSDF file format
#Ref: http://www.nndc.bnl.gov/nndcscr/documents/ensdf/ensdf-manual.pdf
levels = Float64[]
open(fn) do f
    for line in readlines(f)
        length(line) ≥ 19 || continue
        line[6:8] == "  L" && try
            push!(levels, float(line[10:19])) #Most lines have data here
        catch ArgumentError 
            info("Skipped line: $line")
        end
    end
end
x̄=mean(levels)
n=length(levels)
println("The data for $nuclide contains $n energy levels.")
```

Now that we have read in the nuclear energy levels, we now compute their normalized histogram. This histogram is normalized in two ways:

1. The data are rescaled so that the mean is exactly 1.
2. The counts are normalized so that the result is a probability density.

```julia
#Normalized histogram
function normhist(data::AbstractVector, x̄=nothing; nbins = 0)
    n = length(data)
    nbins == 0 && (nbins = int(√(n)))
    x̄ == nothing && (x̄ = mean(data))
    xgrid, counts = hist(data, linspace(minimum(data), maximum(data), nbins))
    xgrid/x̄, counts*x̄/(n*(xgrid[2]-xgrid[1]))
end
```

Now plot the results against Wigner's surmise,

$$W(x) = \frac \pi 2 x \exp\left(-\pi \frac {x^2} 4\right),$$

which is an accurate approximation of the level spacing of the Gaussian unitary ensemble (GUE). The Wigner surmise is accurate for some nuclei (e.g. $^{238}$U) but not others (e.g. $^{208}$Pb).

```julia
xgrid2 = linspace(0, maximum(levels)/x̄, n)
@manipulate for nb=1:2int(√n) begin
    xgrid, counts = normhist(levels, nbins=nb)
    plot(Guide.title("Energy levels for $nuclide (N = $n)"),
        Guide.xlabel("Normalized energy (mean = 1)"), Guide.ylabel("Probability density"),
        layer(x=xgrid2, y=map(x->π/2*x*exp(-π*x^2/4), xgrid2), 
            Geom.line, Theme(default_color=color("red"))),
        layer(x=midpoints(xgrid), y=counts, Geom.bar, Theme(default_color=color("black"))))
end end
```
