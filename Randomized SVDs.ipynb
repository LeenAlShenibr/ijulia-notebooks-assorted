```julia
"""
Randomized matrix norm estimator

Computes a probabilistic upper bound on the norm of a matrix A.

Lemma 4.1 of (Halko, 2011) states (with slight notational change)
that

‖A‖ ≤ α √(2/π) maxᵢ ‖Aωᵢ‖

with probability p=α^(-r).

Inputs:
    A: Matrix whose norm to estimate
    r: Number of matrix-vector products to compute
    p: Probability of upper bound failing (default: 0.05)

Output:
    Estimate of ‖A‖.
"""
function rnorm(A, r::Int, p::Real=0.05)
    α = p^(-1.0/r)
    m, n = size(A)
    Ω = randnn(eltype(A), n, r)
    AΩ = A*Ω
    mx = maximum([norm(sub(AΩ, :, j)) for j=1:r])
    α * √(2/π) * mx
end
```

```julia
"""
Randomized matrix norm estimator using A'A

Computes a probabilistic upper bound on the norm of a matrix A.

The Reference states (with minor change in notation) that 

ρ = √(‖(A'A)ʲω‖/‖(A'A)ʲ⁻¹ω‖)

which is an estimate of the spectral norm of A produced by j
steps of the power method starting with normalized ω, is a lower
bound on the true norm by a factor

ρ ≤ α ‖A‖

with probability greater than 1 - p, p = 4√(n/(j-1))*α^(-2j).

Inputs:
    A: Matrix whose norm to estimate
    j: Number of power iterations to perform (default: 1)
    p: Probability of upper bound failing (default: 0.05)
    At (optional keyword): Transpose of A. Default: A'

Output:
    Estimate of ‖A‖.

Reference:
    Edo Liberty, Franco Woolfe, Per-Gunnar Martinsson, Vladimir Rokhlin and Mark Tygert,
    "Randomized algorithms for the low-rank approximation of matrices",
    Proceedings of the National Academy of Sciences, 104 (51), 2007, 20167-20172,
    doi:10.1073/pnas.0709640104
    Appendix
"""
function rnorms(A, j::Int=1, p::Real=0.05; At = A')
    m, n = size(A)
    α = ((j-1)/n*(p/4)^2)^(-1/(4j))

    Ωold = Ω = randnn(eltype(A), n)
    for i=1:j #Power iterations
        Ω, Ωold = At*(A*Ω), Ω
    end
    ρ = √(norm(Ω)/norm(Ωold))
    α*ρ
end

"""
Randomized Gaussian matrices normalized by column

Input:
    el: element type
    m: number of rows
    n: number of columns or nothing
"""
function randnn(el, m::Int)
    if el <: Real
        Ω = randn(m)
    elseif el <: Complex
        Ω = randn(m) + im*randn(m)
    else
        throw(ValueError("Unsupported element type: $el"))
    end
    Ω/norm(Ω)
end
function randnn(el, m::Int, n::Int)
    if el <: Real
        Ω = randn(m, n)
    elseif el <: Complex
        Ω = randn(m, n) + im*randn(m, n)
    else
        throw(ValueError("Unsupported element type: $el"))
    end
    for i=1:n
        Ω[:, i] /= norm(sub(Ω, :, i))
    end
    Ω
end
```

```julia
"""
Randomized condition number estimator

Inputs:
    A: Matrix whose condition number to estimate.
       Must be square and support premultiply (A*⋅) and solve (A\⋅)
    k: Number of power iterations to run. Default: 1 (Recommended: k ≤ 3)
    p: Probability that estimate fails to hold as an upper bound (Default: 0.05)

Output:
    The interval (x, y) which contains κ(A) with probability 1-p.

Implementation note:
    The reference originally describes this as a computation that can be done by
    computing the necessary number of power iterations given p and the desired accuracy parameter θ=y/x.
    However, these bounds were only derived under the assumptions of exact arithmetic.
    Empirically, k≥4 has been seen to result in incorrect results in that the computed interval does not
    contain the true condition number. We therefore make k an explicitly user-controllable parameter.

Reference:
    @article{Dixon1983,
        author = {Dixon, John D},
        doi = {10.1137/0720053},
        journal = {SIAM Journal on Numerical Analysis},
        number = {4},
        pages = {812--814},
        title = {Estimating Extremal Eigenvalues and Condition Numbers of Matrices},
        volume = {20},
        year = {1983}
    }
"""
function rcond(A, k::Int=1, p::Real=0.05)
    @assert 0<p≤1
    m, n = size(A)
    @assert m==n
    θ = (8n/(π*p^2))^(1/k)
    x = randnn(eltype(A), n)
    for i=1:k
        x = A*x
    end
    y = randnn(eltype(A), n)
    for i=1:k
        y = A\y
    end
    φ = ((x⋅x)*(y⋅y))^(1/(2k))
    (φ, θ*φ)
end
```

```julia
#Algorithm 4.2: adaptive randomized range finder
function rrange(A, r::Integer, ϵ::Real=eps(); maxiter=10)
    m, n = size(A)
    Ω = randn(n, r)
    Y = A*Ω
    Q = zeros(m, 0)
    const tol=ϵ/(10*√(2/π))
    for j=1:maxiter
        Tol = maximum([norm(Y[:,i]) for i=j-1+(1:r)])
        info("Iteration $j: norm = $Tol (target: $tol)")
        Tol > tol || break

        y = sub(Y, :, j)
        if j>1
            Y[:,j]-=Q*Q'y
        end
        q = y/norm(y)
        Q = [Q q]
        ω = randn(n)
        y = A*ω
        y -= Q*Q'y
        Y = [Y y]
        for i = j+(1:(r-1))
            y = sub(Y, :, i)
            Y[:,i] -= q*(q⋅y)
        end
        j==maxiter && warn("Maximum number of iterations reached")
    end
    Q
end
```

```julia
#Algorithm 4.3: randomized power iteration (not recommended)
#q is the number of refining power iterations. The power iteration suppresses the contribution of smaller singular values.
function rrange_pow(A, l::Integer; At=A', p::Integer=0, q::Integer=0, basis=_->qr(_)[1])
    p>=0 || error()
    n = size(A, 2)
    Ω = randn(n,l+p)
    Y = A*Ω
    for iter=1:q #Do some power iterations
        Y=A*At*Y
    end
    Q = basis(Y)
    p==0 ? Q : Q[:,1:l]
end
```

```julia
approx = rrange_pow(A,k)
plotbasis(approx, exact)
```

```julia
#Algorithm 4.4: randomized subspace iteration (recommended)
function rrange_si(A, l::Integer; At=A', p::Integer=0, q::Integer=0, basis=_->qr(_)[1])
    p>=0 || error()
    n = size(A, 2)
    Ω = randn(n,l+p)
    Y = A*Ω
    for iter=1:q #Do some power iterations
        Ỹ=At*Q
        Q̃=basis(Ỹ)
        Y=At*Q̃
        Q=basis(Y)
    end
    Q = basis(Y)
    p==0 ? Q : Q[:,1:l]
end
```

Section 4.6 says that we can use many other $\Omega$s; structured choices can make the matrix product $A\Omega$ faster to compute. They recommend the subsampled random Fourier transform (4.6). Another good choice is random Givens rotations (4.8).

Postmultiplying a matrix $A$ with the subsampled random Fourier transform consists of four steps:

- Scale by a constant real number
- Postmultiply with a diagonal matrix with random elements distributed uniformly on the complex unit circle

```julia
import Gadfly.plot
function plot{T<:Complex}(A::Vector{T}, args...)
    plot(#The unit circle
        layer(x=map(sin,0:π/12:2π), y=map(cos,0:π/12:2π), Geom.line(preserve_order=true)),
        #The individual points
        layer(Geom.point, x=real(A), y=imag(A)), args...)
end
plot(exp(2π*im*rand(200)))
```

- Postmultiply by the discrete Fourier transform. Here we check that multiplication by $F$ uses the correct definition according to the description in (4.6). It turns out that a factor of $\sqrt{n}$ is needed to reconcile Julia's definition using FFTW and the definition the paper wants.

```julia
#Generate the dft matrix
function dftm(n::Integer)
    F=zeros(Complex128, n, n)
    for p=1:n, q=1:n
        F[p,q] = exp(-2π*im*(p-1)*(q-1)/n)/sqrt(n)
    end
    F
end

n=2
a=(randn(n)+im*randn(n))
@test_approx_eq dftm(n)*a fft(a)/√n
a=a'
@test_approx_eq a*dftm(n) fft(a)/√n
m=3
A=randn(m,n)+im*randn(m,n)
@test_approx_eq dftm(m)*A hcat([fft(A[:,i]) for i=1:n]...)/√m
@test_approx_eq A*dftm(n) vcat([fft(A[i,:]) for i=1:m]...)/√n
```

- Postmultiply by a sampling of $l$ columns without replacement from the $n\times n$ identity matrix

```julia
m=6
n=5
l=3
@show randperm(n)
@show Π=randperm(n)[1:l]
@show eye(n)[:,Π]
A=randn(m,n)
@test_approx_eq A*eye(n)[:,Π] A[:,Π]
@test_approx_eq eye(m)[Π,:]*A A[Π,:]
```

Putting the pieces together gives us a definition for postmultiplication with the SFRT.

```julia
function *(A, Ω::srft)
    m, n = size(A)
    B = A*Diagonal(exp(2π*im*rand(n))/√Ω.l)
    B = vcat([fft(A[i,:]) for i=1:m]...) #Factor of √n cancels out
    B[:, randperm(n)[1:Ω.l]]
end

#Algorithm 4.5: fast randomized range finder
function rrange_f(A, l::Integer; p::Integer=0, basis=_->qr(_)[1])
    p>=0 || error()
    n = size(A, 2)
    Ω = srft(n,l+p)
    Y = A*Ω
    Q = basis(Y)
    p==0 ? Q : Q[:,1:l]
end
```

# Stage B: Construction of factorizations

```julia
#Algorithm 5.1: direct SVD
#More accurate
function svdfact(A, Q)
    B=Q'*A
    S=svdfact(B)
    SVD(Q*S[:U], S[:S], S[:Vt])
end
```

```julia
#Algorithm 5.2: SVD via row extraction
#Faster but less accurate
#Recommended input of Q is Y=A*Ω
function svdfact_re(A, Q)
    X, J = id(Q)
    R′, W′= qr(A[J, :])
    Z = X*R′
    S=svdfact(Z)
    SVD(S[:U], S[:S], S[:Vt]*W′)
end
```

```julia
#Algorithm 5.3: direct eigenvalue decomposition
function eigfact(A::Hermitian, Q)
    B = Q'A*Q
    E = eigfact!(B)
    Eigen(E[:values], Q*E[:vectors])
end
```

```julia
#Algorithm 5.4: eigenvalue decomposition via row extraction
function eigfact_re(A::Hermitian, Q)
    X, J = idfact(Q)
    F = qrfact!(X)
    V, R = F[:Q], F[:R]
    Z=R*A[J, J]*R'
    E=eigfact(Z)
    Eigen(E[:values], V*E[:vectors])
end
```

```julia
#Algorithm 5.5: eigenvalue decomposition via Nyström method
function eigfact_ny_posdef(A, Q)
    B1=A*Q
    B2=Q'*B1
    C=cholfact!(B2)
    F=B1/C
    S=svdfact!(F)
    Eigen(S[:S].^2, S[:U])
end
```

```julia
#Algorithm 5.6: one-pass eigenvalue decomposition
#Can be improved by oversampling
#Y = Q*Q'Ya
function eigfact_onepass(A::Hermitian, Ω)
    Y=A*Ω
    B=(Q'Y)\(Q'Ω)
    E=eigfact!(B)
    Eigen(E[:values], Q*E[:vectors])
end

function eigfact_onepass(A, Ω, Ω̃; At=A')
    Y=A *Ω; Q = qrfact!(Y)[:Q]
    Ỹ=At*Ω; Q̃ = qrfact!(Ỹ)[:Q]
    #Want least-squares solution to (5.14 and 5.15)
    B=(Q'Y)\(Q̃'Ω)
    B̃=(Q̃'Ỹ)\(Q'Ω̃)
    #Here is a very very very hacky way to solve the problem
    B=0.5(B + B̃')
    E=eigfact!(B)
    Eigen(E[:values], Q*E[:vectors])
end
```

# Ancillary routines

```julia
#Algorithm 2.2: modified Gram-Schmidt and
#Algorithm 2.3: classical Gram-Schmidt
#
# References:
# - Åke Björck, Numerics of Gram-Schmidt orthogonalization,
#      doi: 10.1016/0024-3795(94)90493-6
#
#Q: Destination of the orthogonal basis. Default is A (overwrite A)
#   The default only makes sense if A is a dense matrix, and a
#   different destination for Q should be specified otherwise
#modified: whether to use modified Gram-Schmidt (true; default)
#          or classical Gram-Schmidt (false)
#tol: if positive, performs a partial decomposition in the
#     style of Businger and Golub. After a new column is
#     processed, checks the Frobenius norm of the remaining
#     columns and terminates if it is smaller than tol.
#     Default is 0 (doesn't check)
#iters: if >1, performs additional reorthogonalization sweeps
#       Gram-Schmidt. It is rare that iters>2 is needed.
#       Default is 1 (don't reorthogonalize)
#normalize: whether or not to compute an orthonormal basis (default).
#           If false, the computed basis is not explicitly normalized;
#           instead R is unit upper triangular. The unnormalized variant
#           is square root-free.
function qr!(A; Q=A, modified::Bool=true, tol::Real=0,
        iters::Integer=1, normalize::Bool=true)
    n=size(A,2)
    #A gets overwritten by Q
    R=eye(eltype(A),n,n)
    kstop=1
    for k=1:n
        kstop=k
        if modified #Run modified Gram-Schmidt
            for i=1:k-1 
                R[i,k] = dot(Q[:,i], A[:,k])
                Q[:,k]= A[:,k] - R[i,k]*Q[:,i]
            end
        else #Classical Gram-Schmidt
            R[1:k-1,k]=Q[:,1:k-1]'*A[:,k]
            Q[:,k]= A[:,k] - Q[:,1:k-1]*R[1:k-1,k]
        end
        if normalize
            R[k,k]=norm(A[:,k])
            Q[:,k]/=R[k,k]
        end
        #If computing a partial factorization, check if conve
        tol>0 && normfro(A[:,k+1:n])<tol && break 
    end
    if kstop<n #Truncate if necessary
        Q=Q[:,1:kstop]
        R=R[1:kstop,1:kstop]
    end
    if iters<=1 #Reorthogonalize?
        Q, Triangular(R, :U, !normalize)
    else
        Q, Rs = qr_mgs!(Q, iters=iters-1, normalize=normalize)
        Q, Rs*Triangular(R, :U, !normalize)
    end
end
```

```julia
function id(A)
    m, n = size(A)
    
    #Step 1
    QRF = qrfact(A, pivot=true) #Should use pivoted Gram-Schmidt
    PR, Q, R = QRF[:P], QRF[:Q], QRF[:R]
    @show k = min(m, n) #Actually the cutoff from the incomplete qr
    
    #Step 2
    T = R[1:k, 1:k] \ R[1:k, k+1:end]
    @assert size(T) == (k, n-k)

    A_CS = (A*PR)[:, 1:k]
    @assert size(A_CS) == (m, k)

    #A≈A_CS*[eye(k) T]*PR'
    #@assert norm(A - (A_CS*[eye(k) T]*PR')) < √(eps())

    #Step 3
    QRF_CS = qrfact(A_CS', pivot=true) #Should use pivoted Gram-Schmidt
    PLt, Q_CSt, R_CSt = QRF_CS[:P], QRF_CS[:Q], QRF_CS[:R]

    #Step 4
    St = R_CSt[1:k, 1:k] \ R_CSt[1:k, k+1:end]
    @assert size(St) == (k, m-k)
    
    #A_CS = PLt'*[eye(k); St']*(A_CS'*PLt)'[1:k, :]
    #@assert norm(A_CS - PLt'*[eye(k); St']*(A_CS'*PLt)'[1:k, :] ) < √(eps())
    
    PLt', St', (A_CS'*PLt)'[:, 1:k], T, PR'
end
```

```julia
#Randomized range finder
#Algorithm 4.1: randomized range finder (not recommended)
#A must support size(A, 2) and premultiply
#basis is the algorithm to compute the orthogonal basis Q
#They recommend Gram-Schmidt with double orthogonalization
#Ωf: choice of test basis generating function
#    A safe choice is x->randn(n,l+p)
#    A faster choice is x->srft(n,l+p)
#p is the oversampling parameter
#q is the number of refining power iterations. The power iteration suppresses the contribution of smaller singular values.
function rrange(A, l::Integer; At=A', Ωf = x->randn(n,l+p),
        p::Integer=0, q::Integer=0, basis=_->qr(_)[1])
    p>=0 || error()
    n = size(A, 2)
    Ω = Ωf()
    Y = A*Ω
    for iter=1:q #Do some power iterations
        Y=A*At*Y
    end
    Q = basis(Y)
    p==0 ? Q : Q[:,1:l]
end

#Algorithm 4.2: adaptive randomized range finder
function rrange(A, r::Integer, ϵ::Real; maxiter=10)
    m, n = size(A)
    Ω = randn(n, r)
    Y = A*Ω
    Q=zeros(m, 0)
    @show const tol=ϵ/(10*√(2/π))
    for j=1:maxiter
        Tol = maximum([norm(Y[:,i]) for i=j-1+(1:r)])
        @show Tol
        Tol > tol || break
        j>1 && (Y[:,j]=(I-Q*Q')*Y[:,j])
        q = Y[:,j]/norm(Y[:,j])
        Q = [Q q]
        ω = randn(n)
        y = (I-Q*Q')*A*ω
        Y = [Y y]
        for i = j+(1:(r-1))
            Y[:,i] -= q*dot(q, Y[:,i])
        end
        j==maxiter && warn("Maximum number of iterations reached")
    end
    Q
end

#Algorithm 4.4: randomized subspace iteration (recommended)
function rrange_si(A, l::Integer; At=A', p::Integer=0,
        q::Integer=0, basis=_->qr(_)[1])
    p>=0 || error()
    n = size(A, 2)
    Ω = randn(n,l+p)
    Y = A*Ω
    for iter=1:q #Do some power iterations
        Ỹ=At*Q
        Q̃=basis(Ỹ)
        Y=At*Q̃
        Q=basis(Y)
    end
    Q = basis(Y)
    p==0 ? Q : Q[:,1:l]
end
```
